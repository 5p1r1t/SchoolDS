# Собрать информацию о всех строящихся объектах на сайте "наш.дом.рф" (не все, limit_ = 100 объектов)

```python
object_data = [requests.get(f'https://наш.дом.рф/сервисы/api/object/{x}').json().get('data') for x in objids]
# len(object_data)
```

```python
object_data[0]
```

# Cохранить ее в pandas dataframe

```python
df = pd.json_normalize(object_data)
```

```python
df
```

# Cохранить pandas dataframe в excel

```python

df.to_excel("Task1_df_to_excel.xlsx")
```

# Cохранить pandas dataframe в pickle

```python
df.to_pickle("Task1_df_to_pickle.pkl")
```

# Cохранить pandas dataframe в БД

```python
df1 = df.drop(columns=['photoRenderDTO','objectTransportInfo','metro.colors'])
```

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite://', echo=False)


df1.to_sql('objects', con=engine)



engine.execute("SELECT * FROM objects").fetchall()
```
